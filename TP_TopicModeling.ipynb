###Partie I : Implémenter le modèle de topic modeling LDA sur le dataset 20newsgroups afin d'identifier les sujets (topics) présents dans les documents.

```python
# prompt: 1. Importation des bibliothèques

import numpy as np
import pandas as pd
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
# prompt: 2. Chargement du dataset
newsgroups_data = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))
```

```python
# Création d'un DataFrame pandas
df = pd.DataFrame({'text': newsgroups_data.data})
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

# Tokenization
df['text'] = df['text'].apply(word_tokenize)

# Lemmatization
lemmatizer = WordNetLemmatizer()
df['text'] = df['text'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])

# Stopwords removal
stopwords_list = stopwords.words('english')
df['text'] = df['text'].apply(lambda x: [word for word in x if word not in stopwords_list])
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Convert the preprocessed text back to string
df['text'] = df['text'].str.lower()

# Create an instance of CountVectorizer
vectorizer = CountVectorizer(max_features=1000, stop_words='english')

# Fit and transform the preprocessed text data
X = vectorizer.fit_transform(df['text'])
```

```python
from sklearn.decomposition import LatentDirichletAllocation

num_topics = 10  # Choose the desired number of topics
lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)
lda.fit(X)
```

```python
# Afficher les mots-clés associés à chaque topic
feature_names = vectorizer.get_feature_names_out()
for topic_idx, topic in enumerate(lda.components_):
 print(f"Topic #{topic_idx + 1}:")
 print([feature_names[i] for i in topic.argsort()[-10:]])
 print()
# Assigner chaque document au topic dominant
df['dominant_topic'] = lda.transform(X).argmax(axis=1)
```

#Partie II : Implémenter la méthode LSA sur le dataset 20newsgroups afin d'extraire les thèmes (topics) présents dans les documents

```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
```

```python
# Téléchargement du dataset
newsgroups_data = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))
# Création d'un DataFrame pandas
df = pd.DataFrame({'text': newsgroups_data.data})
```

```python


# Tokenization
df['text'] = df['text'].apply(word_tokenize)

# Lemmatization
lemmatizer = WordNetLemmatizer()
df['text'] = df['text'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])

# Stopwords removal
stopwords_list = stopwords.words('english')
df['text'] = df['text'].apply(lambda x: [word for word in x if word not in stopwords_list])
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Convert the preprocessed text back to string
df['text'] = df['text'].str.lower()
vectorizer = TfidfVectorizer(max_features=1000, stop_words='english')
X_tfidf = vectorizer.fit_transform(df['text'])
```

```python
num_topics = 10 # Choisissez le nombre de thèmes souhaité
lsa = TruncatedSVD(n_components=num_topics, random_state=42)
lsa.fit(X_tfidf)
```

```python
# Afficher les mots-clés associés à chaque thème
feature_names = vectorizer.get_feature_names_out()
for topic_idx, topic in enumerate(lsa.components_):
 print(f"Thème #{topic_idx + 1}:")
 print([feature_names[i] for i in topic.argsort()[-10:]])
 print()
# Assigner chaque document au thème dominant
df['dominant_topic'] = lsa.transform(X_tfidf).argmax(axis=1)
```
